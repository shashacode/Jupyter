```python
co2data["CO2 emissions (Kilo-tonnes)"] = (co2data["Annual CO₂ emissions (tonnes )"]/1000000).round(2)
#d emissions are converted into kilo-tonnes
```

```python
#filter only required columns and omit the annual co2 emissions (tonnes)
co2data = co2data[["Entity", "Code", "Year", "CO2 emissions (Kilo-tonnes)"]]
```

```python
target_country_data = co2data[co2data["Entity"] == "Afghanistan"] #filter required country only data

#get rolling diff for data
rolling_differences = target_country_data["CO2 emissions (Kilo-tonnes)"].diff()

#create new empty dataframe and add year from target country n rolling diff calculated
new_df = pd.DataFrame()
new_df["Year"] = target_country_data["Year"].iloc[1:]
new_df["Differences"] = rolling_differences
new_df
```

```python
target_country_data = co2data[co2data["Entity"] == "Africa"] #filter required country only data

#get rolling diff for data
rolling_differences = target_country_data["CO2 emissions (Kilo-tonnes)"].diff()

#create new empty dataframe and add year from target country n rolling diff calculated
new_df = pd.DataFrame()
new_df["Year"] = target_country_data["Year"].iloc[1:]
new_df["Differences"] = rolling_differences
new_df

year_of_max_increase = new_df[new_df.Differences == new_df.Differences.max()]
year_of_max_increase
```
