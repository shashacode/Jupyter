```python
import matplotlib.pyplot as plt
import numpy as np
#%pip install seaborn
import seaborn as sns
plt.style.use("ggplot")
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
z = [45,47,52,55,50,51,60]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('Night')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x, y, color = 'green', alpha =1, linewidth = 1, label = "Osun")
plt.plot(x,z,color = 'blue', alpha = 10, linewidth =1, label = 'Lagos')
plt.legend()
plt.show()
```

```python
value_nigeria =[58, 82, 118, 155, 170, 291, 392, 451, 972, 1089, 1093,1125,200]
value_sudan = [35.8, 45.8, 54.5, 53.1, 65.6, 67.3, 68.1, 72.0, 82.1, 97.1, 95.5, 117.4,200]
year = [1980, 1985, 1990, 1995, 2000, 2005, 2010, 2012, 2014, 2015, 2016, 2017,2018]
print(len(value_nigeria), len(value_sudan))
```

```python
plt.xlabel('Year')
plt.ylabel('GDP - Billion (USD)')
plt.title('GDP-TIME TREND')

plt.plot(year, value_nigeria, color = 'green', alpha =1, linewidth = 1, label = "Nigeria")
plt.plot(year,value_sudan,color = 'blue', alpha = 10, linewidth =1, label = 'Sudan')
plt.plot(year,list(reversed(value_sudan)),color = 'red', alpha = 1, linewidth =1, label = 'Ghana')
plt.legend()
plt.show()
```

```python
plt.xlabel('Year')
plt.ylabel('GDP - Billion (USD)')
plt.title('GDP-TIME TREND')

plt.bar(year, value_nigeria, color = 'green', alpha =1, linewidth = 1, label = "Nigeria")
plt.bar(year,value_sudan,color = 'blue', alpha = 1, linewidth =1, label = 'Sudan')
# plt.plot(year,list(reversed(value_sudan)),color = 'cyan', alpha = 10, linewidth =1, label = 'Ghana')
plt.legend()
plt.show()
```

```python
temp_axis = np.arange(len(year))
```

```python
plt.xlabel("Year")
plt.ylabel('GDP-Billion (USD)')
plt.title('GDP-TIME TREND')

bar_width = 0.2

plt.bar(temp_axis -0.3, value_nigeria, width=bar_width, label = "Nigeria")
plt.bar(temp_axis,value_sudan,width=bar_width, label = 'Sudan')
plt.bar(temp_axis + 0.3, list(reversed(value_nigeria)), width=bar_width, label = 'Ghana')

plt.xticks(temp_axis, year) #replaces d current arange tick values to actual year values
plt.legend()
plt.show()
```

```python
plt.xlabel("Year")
plt.ylabel('GDP-Billion (USD)')
plt.title('GDP-TIME TREND')

bar_width = 0.4

plt.bar(year, value_nigeria, width=bar_width, label = "Nigeria")
plt.bar(year,value_sudan,width=bar_width, label = 'Sudan', bottom = value_nigeria)
plt.bar(year, list(reversed(value_nigeria)), width=bar_width, label = 'Ghana', 
        bottom = np.array(value_nigeria) + np.array(value_sudan))

# plt.xticks(temp_axis, year) #replaces d current arange tick values to actual year values
plt.legend()
plt.show()
```
