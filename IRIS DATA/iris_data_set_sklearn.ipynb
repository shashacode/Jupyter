```python
import pandas as pd
# import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
# Create KNN classifier
# knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
data = pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
data.describe()
# data = data.drop("Id", axis = 1)
# data.head()
```

```python
sns.pairplot(data, hue = "Species")
```

```python
ml_features = data[["PetalLengthCm","PetalWidthCm" ]]
ml_labels = data["Species"]
```

```python
x_train, x_test, y_train, y_test = train_test_split( ml_features, ml_labels, test_size = 0.1, random_state = 1) #cross validation
```

```python
model = KNeighborsClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
output_data = pd.DataFrame({})
output_data["Actual"] = y_test
output_data["Predicted"] = y_pred
```

```python
output_data
```

```python
accuracyscore = metrics.accuracy_score
```

```python
accuracyscore(output_data["Actual"], output_data["Predicted"])
```
